```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score
```

```python
data_url ='./data giua ki 2.csv'
df=pd.read_csv(data_url, encoding='unicode_escape')
df
```

```python
win_counts = df.groupby('team')['wins'].sum().reset_index()
win_counts = win_counts.sort_values(by='wins', ascending=False)

top_5_teams = win_counts.head(5)
color = ['mediumseagreen']

plt.bar(top_5_teams['team'], top_5_teams['wins'],color=color)
plt.xticks(rotation=50)

plt.title('Top 5 Teams with Most Wins (2006-2018)')
plt.xlabel('Teams')
plt.ylabel('Wins')

```

```python
sns.regplot(data=df, x='penalty', y='pen_goals',color='r', marker='*')

plt.title('Correlation between penalty and pen_goals')
plt.xlabel('Penalty')
plt.ylabel('Pen_Goals')
plt.show()
```
